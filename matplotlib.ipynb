```python
x = np.linspace(start=0,stop=10,num=50)
plt.plot(x,np.sin(x), label = 'SINE Curve')

plt.xlabel('x', fontsize=15, color='green')
plt.ylabel('sin(x)', fontsize=15, color='green');

plt.title('Basics of PLotting')

plt.xlim(1,5)

plt.tick_params(axis='y',
               color='red',
               labelcolor='blue',
               labelsize='xx-large')

plt.tick_params(axis='x',
               bottom=True,
               labelbottom=True)



plt.legend(loc=0);
```

```python
x = np.linspace(start=0,stop=10,num=50)
plt.plot(x,np.sin(x), label='sine curve', color='red')
plt.plot(x,np.cos(x), label='cosine curve', color = 'green')

plt.legend()
plt.title('Learning Basic plots')
```
