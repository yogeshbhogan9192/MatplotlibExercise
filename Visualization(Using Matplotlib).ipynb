```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
matplotlib.get_backend()
```

```python
plt.plot ([2,4,6,8],
         [4,8,12,16], color = 'purple')
plt.xlabel('X-Axis', fontsize = 20 , color = 'red')
plt.ylabel('Y-Axis', fontsize = 20, color = 'green')
```

```python
x = np.linspace(start = 0 , stop = 10 , num = 50)
```

```python
x.shape
```

```python
plt.plot(x, np.sin(x))

plt.title('Basic Graph', fontsize = 30 , color = 'orange')

plt.xlabel('X-Axis', fontsize = 15, color ='red')
plt.ylabel('Y-Axis', fontsize = 15, color ='red')


plt.tick_params(axis = 'y',
               color = 'red',
               labelcolor = 'green',
               labelsize='xx-large')


plt.tick_params(axis = 'x',
               color = 'red',
               labelcolor = 'blue',
               labelsize='xx-large')

#plt.xlim(0,5)
#plt.ylim(0.8,1.0)
```

```python
x = np.linspace(start=0,stop=10,num=50)
plt.plot(x,np.sin(x), label='sine curve', color='red')
plt.plot(x,np.cos(x), label='cosine curve', color = 'green')

plt.legend()
plt.title('Learning Basic plots')
```

```python
np.random.seed(0)
random_array = np.random.randn(30)
plt.plot(random_array, color='orange', linestyle = '-.',lw = 4)
```

```python

#import pandas_image as im

fig = plt.figure(figsize = (10,12))

ax1 = plt.subplot2grid((2,3),(0,0))
ax1.plot([2,4,6,8],[4,8,12,16])

ax2 = plt.subplot2grid((2,3),(0,1)) #2,3 matrix will have 6 axes(6quadrants)
ax2.plot(x,np.cos(x), label='cosine curve', color = 'green')

ax3 = plt.subplot2grid((2,3),(0,2),rowspan = 2)
ax3.plot(x,np.sin(x), label='sine curve', color = 'green')

ax4 = plt.subplot2grid((2,3),(1,0),colspan = 2)
ax4.plot(x,np.exp(x), label='exp curve', color = 'green')

plt.tight_layout()
```
